#Gradient Boosting Regression


```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import OneHotEncoder
```

```python

X = np.array([
    [160, "Red",   0],  # Female, Red
    [170, "Blue",  1],  # Male, Blue
    [165, "Green", 0],  # Female, Green
    [180, "Black", 1],  # Male, Black
    [175, "Blue",  1],  # Male, Blue
    [155, "Red",   0],  # Female, Red
    [168, "Green", 0],  # Female, Green
    [185, "Black", 1],  # Male, Black
    [178, "Blue",  1],  # Male, Blue
    [162, "Red",   0],  # Female, Red
    [172, "Green", 0],  # Female, Green
    [177, "Black", 1],  # Male, Black
    [158, "Red",   0],  # Female, Red
    [182, "Blue",  1],  # Male, Blue
    [167, "Green", 0],  # Female, Green
    [188, "Black", 1],  # Male, Black
    [163, "Red",   0],  # Female, Red
    [176, "Blue",  1],  # Male, Blue
    [169, "Green", 0],  # Female, Green
    [183, "Black", 1],  # Male, Black
], dtype=object)

y = np.array([
    55, 72, 60, 85, 78,
    50, 58, 90, 80, 54,
    62, 83, 52, 88, 59,
    92, 56, 77, 61, 87
])
```

```python
encoder = OneHotEncoder(sparse_output=False)
X_encoded = encoder.fit_transform(X[:, 1].reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

## Now from Scratch

```python
def gradient_boosting_regression(X, y, n_estimators=100, lr=0.1, max_depth=3):
  avg = np.mean(y_train)
  residuals = np.subtract(y_train, avg)
  trees = []
  for i in range(n_estimators):
    residual_tree = DecisionTreeRegressor(max_depth=max_depth)
    trees.append(residual_tree)
    residual_tree.fit(X, residuals)
    residuals = residuals - lr * residual_tree.predict(X)
  return trees, avg
```

```python
def predict(X, trees, avg, lr=0.1):
    y_pred = np.full(X.shape[0], avg)
    for tree in trees:
        y_pred += lr * tree.predict(X)
    return y_pred
```

```python
trees, avg = gradient_boosting_regression(X_train, y_train)
y_pred = predict(X_test, trees, avg)

error =0
for i in range(len(y_pred)):
  error += (y_pred[i] - y_test[i])**2
error = error/len(y_pred)
print("MSE:", error)
```
